## 3. cliet에서 server로 보내는 request packet modify하기

[참조링크](http://stackoverflow.com/questions/9063583/python-twisted-proxy-how-to-intercept-packets)


```python
class ClientProxy(proxy.Proxy):
    def dataReceived(self, data):

        print "Before : "
        #Modify the data
        print data
        
        print "After : "
        data = string.replace(data, 'beatles', 'ahroobe')
        print data
        #perform the default functionality on modified data 
        return proxy.Proxy.dataReceived(self, data)

class ProxyFactory(http.HTTPFactory):
    protocol=ClientProxy

factory = ProxyFactory()
reactor.listenTCP(8080, factory)
reactor.run()
```

* 초창기 log 부분

이와 같이 data가 print되는데, 문제는
`(unable to obtain traceback): 'unicode' does not have the buffer interface` 라는 에러가 뜬다. 그리고 실제로 request가 보내지지 않아서 Wireshark를 보면 HTTP통신이 제대로 되지 않는다.
data를 ""인 빈 스트링으로 적용하고 실행했을때는 그런 에러가 발생하지는 않으나 여전히 request가 보내지지 않는다.
